### Part 5: Experiments
To address the task of predicting the price of Rolex Submariner watches, several data preprocessing steps and learning algorithms were employed. These steps were carefully chosen based on the nature of the data and the problem at hand.

- Removal of Non-Submariner Type: As there were three Submariner types in the dataset, it was observed that Rolex only produces two types: one with a date and one without a date. To align with this information, the last Submariner type with only 164 observations was removed from the dataset.

- Handling Missing Values: The "year of production" feature exhibited a relatively large number of missing values. To capture potential differences related to missing values, a missing value indicator ("year_of_production_missing") was created as a new binary variable. Missing values in the "year of production" feature were then imputed with the mean value of the available data.

- Removal of Irrelevant Features: The "location" variable, representing the geographic location associated with each observation, was deemed irrelevant for the specific prediction task of price estimation. It was subsequently removed from the dataset to simplify the analysis and avoid potential overfitting issues.

- Selection of Specific Categories: The "case diameter" variable was narrowed down to include only the three most frequent categories ("40 mm," "41 mm," and "41 x 40 mm"). This selection was based on the frequency distribution, with these categories accounting for a significant portion of the observations. By reducing the number of distinct values and potential outliers, the analysis and modeling process were simplified.

- Combination of Categories: The "condition" variable was simplified by combining similar categories. The categories "Very good" and "Unworn" were merged into the "Good" category, while the "Fair" category was relabeled as "Poor." This consolidation reduced redundancy and provided a clearer representation of the overall condition of the Submariner watches.

- Transformation of Reference Number: The "reference number" variable underwent transformation to remove non-numeric characters and group low count categories as "Other." This simplification improved interpretability and focused on essential information.

- Removal of Redundant Features: The "ad name" variable, providing similar information to the "reference number" variable, was removed to avoid redundancy. Additionally, the "movement" variable, which had only two instances for the "Manual winding" category, was removed as it did not contribute significantly to the prediction model.


For the learning algorithms, the following regression models were used:

- Linear Regression: Linear regression was chosen as a baseline model due to its simplicity and interpretability. It assumes a linear relationship between the features and the target variable, making it a suitable choice for capturing straightforward patterns in our data.

- Decision Trees: Decision trees were selected for their ability to handle both numeric and categorical features effectively. They can capture non-linear relationships and interactions between features. Decision trees are also known for their interpretability, as the resulting tree structure can be easily understood and visualized. This makes them useful for identifying important features in predicting the price of Rolex Submariner watches.

- Random Forest: Random forests were chosen as an ensemble method to reduce overfitting and improve prediction accuracy. By combining multiple decision trees, random forests can capture complex relationships and handle high-dimensional datasets. They are robust to outliers and can effectively handle categorical features, making them suitable for the mixed-type features present in the dataset.

- Gradient Boosting: Gradient boosting is another ensemble technique that sequentially builds weak learners to minimize the residual errors. It can handle complex relationships and capture interactions between features. Gradient boosting has the advantage of being able to improve model performance by focusing on the residuals of previous models, and potentially enhancing the accuracy of price predictions.

- Baseline (Dummy): The dummy regressor serves as a reference point for evaluating the performance of more sophisticated models. It predicts the mean  value of our target variable and provides a benchmark against which other models can be compared. It helps assess whether the chosen models are performing significantly better than a simple baseline.

We can see that "year of production" feature has the most missing values. Because there is a reletively large number of missing values in this feature, we decided to create a missing value indicator for it. Instead of removing or imputing missing values, I create a new binary variable that indicates whether "year of production" is missing or not. This can help capture potential differences in the data related to missing values and be used as a separate feature during modeling.

In addition, I will replace missing values for "year of production" with the mean. In my feature selection, I will see which variable is most important.

For the remaining missing values, because there are relatively few compared to the overall size of the dataset, I will remove them.

```python
# Create a missing value indicator for "year of production"
submariner_data['year_of_production_missing'] = submariner_data['year of production'].isnull().astype(int)

# Calculate the mean of "year of production" excluding missing values
mean_year_of_production = submariner_data['year of production'].mean()

# Impute missing values with the mean
submariner_data['year of production'] = submariner_data['year of production'].fillna(mean_year_of_production)

# Remove rows with missing values
submariner_data = submariner_data.dropna()

# Print the number of rows and columns
print("Number of rows:", submariner_data.shape[0])
print("Number of columns:", submariner_data.shape[1])
```

The "location" variable does not seem to be beneficial in my prediction project so I will remove it from my dataset.  The "location" variable represents the geographic location associated with each observation, but for my specific prediction task, the location may not directly impact the target variable. By removing irrelevant variables like "location," I can simplify the analysis and model-building process. Additionally, the "location" variable exhibits high cardinality, with 1234 unique categories, which can lead to challenges in modeling, such as overfitting and increased computational complexity.

```python
submariner_data = submariner_data.drop('location', axis=1)
```

Selecting only the three categories, namely "40 mm", "41 mm", and "41 x 40 mm", for the "case diameter" variable in the submariner_data DataFrame is a good idea. This is because by narrowing down the selection to these specific categories, we are focusing on the most frequent and prominent case diameter sizes. These three categories represent a significant portion of the observations, with "40 mm" accounting for 66.3% of the total, "41 mm" representing 21.9%, and "41 x 40 mm" accounting for 5.7%. Additionally, selecting these categories allows us to simplify the analysis by reducing the number of distinct values and potential outliers in the "case diameter" variable.

```python
selected_cases = ['40 mm', '41 mm', '41 x 40 mm']
submariner_data = submariner_data[submariner_data['case diameter'].isin(selected_cases)]
```

It is also a good idea to combine the categories in the "condition" variable of the submariner_data DataFrame into three categories ("Good", "New", and "Poor") because it simplifies the variable, reduces redundancy, and provides a clearer representation of the overall condition of the Submariner watches.

```python
# Combine categories in the "condition" variable
submariner_data['condition'] = submariner_data['condition'].replace({'Very good': 'Good',
                                                                     'Unworn': 'New',
                                                                     'Fair': 'Poor'})
```

We need to transform the "Reference Number" variable. It removes non-numeric characters, focuses on the essential information, groups low count categories as "Other," and standardizes the representation. This simplifies the variable and improves interpretability

```python
# Extract the numeric part from the "Reference Number" column
submariner_data['reference number'] = submariner_data['reference number'].str.extract(r'(\d+)')

# Calculate the count of each distinct reference number
reference_number_counts = submariner_data['reference number'].value_counts()

# Identify reference numbers with a count of 10 or below
reference_numbers_to_group = reference_number_counts[reference_number_counts <= 20].index

# Replace those reference numbers with "Other"
submariner_data.loc[submariner_data['reference number'].isin(reference_numbers_to_group), 'reference number'] = 'Other'
```

Next, we remove "ad name" variable as the information that it gives is the same as the the "reference number" variable so we only keep one of the two. Also, "movement" variable consists of two categories Automatic and Manual winding, but the later only has 2 instances, which doesnt add much value to our prediction model, if we remove this category we are left with only one category so it's best to remove the feature.

```python
submariner_data = submariner_data.drop(["ad name", "movement"], axis=1)
```

# Categotrical Transformations

```python
import pandas as pd

# One-hot encode Case Material, Condition, Scope Of Delivery, and Case Diameter
case_material_encoded = pd.get_dummies(submariner_data['case material'], prefix='CaseMaterial')
condition_encoded = pd.get_dummies(submariner_data['condition'], prefix='Condition')
scope_encoded = pd.get_dummies(submariner_data['scope of delivery'], prefix='DeliveryScope')
case_diameter_encoded = pd.get_dummies(submariner_data['case diameter'], prefix='CaseDiameter')

# Assign binary values to Model (0 for "Rolex Submariner Date" and 1 for "Rolex Submariner (No Date)")
submariner_data['model'] = (submariner_data['model'] == "Rolex Submariner (No Date)").astype(int)

# Convert Reference Number to numeric
submariner_data['reference number'] = pd.to_numeric(submariner_data['reference number'], errors='coerce')


# Concatenate the encoded features back to the DataFrame
submariner_data_encoded = pd.concat([submariner_data, case_material_encoded, condition_encoded,
                                     scope_encoded, case_diameter_encoded], axis=1)

# Drop the original categorical features
submariner_data_encoded.drop(["case material", "condition", "scope of delivery", "case diameter"],
                             axis=1, inplace=True)

# Print the encoded DataFrame
print(submariner_data_encoded)
```

```python
import pandas as pd
from scipy.stats import spearmanr, pointbiserialr

# Drop rows with missing values
submariner_data_encoded.dropna(inplace=True)

param = []
correlation = []
abs_corr = []

for c in submariner_data_encoded.columns:
    if c != "model":
        if len(submariner_data_encoded[c].unique()) <= 2:
            corr = spearmanr(submariner_data_encoded['model'], submariner_data_encoded[c])[0]
        else:
            corr = pointbiserialr(submariner_data_encoded['model'], submariner_data_encoded[c])[0]
        param.append(c)
        correlation.append(corr)
        abs_corr.append(abs(corr))

param_df = pd.DataFrame({'correlation': correlation, 'parameter': param, 'abs_corr': abs_corr})
param_df = param_df.sort_values(by='abs_corr', ascending=False)
param_df = param_df.set_index('parameter')

print(param_df)
```

# Performance Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.dummy import DummyRegressor

reg_linear = LinearRegression()
reg_tree = DecisionTreeRegressor()
reg_forest = RandomForestRegressor()
reg_gb = GradientBoostingRegressor()
reg_dummy = DummyRegressor()

regressors = {
    reg_linear: 'Linear Regression',
    reg_tree: 'Decision Trees',
    reg_forest: 'Random Forest',
    reg_gb: 'Gradient Boosting',
    reg_dummy: 'Baseline (Dummy)'
}
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# train function fits the model and returns evaluation metrics
def train(algo, name, X_train, y_train, X_test, y_test):
    algo.fit(X_train, y_train)
    y_pred = algo.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"--------------------------------------------{name}---------------------------------------------------")
    print(f"Mean Squared Error for {name}: {mse:.4f}")
    print(f"R-squared Score for {name}: {r2:.4f}")
    return y_test, y_pred, mse, r2

def eval_results(y_test, y_pred):
    print("Evaluation Results")
    print("===================")
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"Mean Squared Error: {mse:.4f}")
    print(f"R-squared Score: {r2:.4f}")

def main(regressors, X_train, y_train, X_test, y_test):
    mse_scores = []
    r2_scores = []
    for algo, name in regressors.items():
        y_test_train, y_pred, mse_score, r2_score = train(algo, name, X_train, y_train, X_test, y_test)
        eval_results(y_test_train, y_pred)
        mse_scores.append(mse_score)
        r2_scores.append(r2_score)
    return mse_scores, r2_scores
```

```python
import matplotlib.pyplot as plt

def rank_models(mse_scores, r2_scores, models):
    ranked_models = sorted(zip(mse_scores, r2_scores, models), key=lambda x: (x[0], -x[1]))
    print("Model Performance Ranking")
    print("=========================")
    for i, (mse, r2, model) in enumerate(ranked_models, start=1):
        print(f"Rank {i}: {model} (MSE: {mse:.4f}, R-squared: {r2:.4f})")

def plot_results(y_test, y_pred):
    plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.title('Actual vs Predicted Price')
    plt.show()

mse_scores, r2_scores = main(regressors, X_train, y_train, X_test, y_test)
rank_models(mse_scores, r2_scores, regressors.values())
```

```python
from yellowbrick.regressor import PredictionError
# Iterate over the models and generate the PredictionError visualizer
for regressor, model_name in regressors.items():
    # Fit the model on the training data
    regressor.fit(X_train, y_train)

    # Instantiate the PredictionError visualizer with the trained model
    visualizer = PredictionError(regressor)

    # Evaluate the model on the test data
    visualizer.score(X_test, y_test)

    # Set the title of the plot to the model name
    visualizer.title = model_name

    # Show the plot
    visualizer.show()
```
